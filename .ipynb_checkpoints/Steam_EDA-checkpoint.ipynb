## ¿Cuál es el género de videojuegos más vendido?

```python
df.columns
```

```python
#Se definen los géneros presentes en el DataFrame en una variable
generos = ['action', 'adventure', 'rpg', 'simulation', 'strategy', 'racing']
```

```python
# Se crea una lista para almacenar las filas
ventas_por_genero = []

# Se calculan las ventas totales por género
for genero in generos:
    total_ventas = df[df[genero] == True]['sells'].sum()  # Sumar ventas donde el género es True
    ventas_por_genero.append({'genero': genero, 'ventas': total_ventas})

# Covertir la lista a un DataFrame
ventas_por_genero_df = pd.DataFrame(ventas_por_genero)

# Género más vendido
genero_mas_vendido = ventas_por_genero_df.loc[ventas_por_genero_df['ventas'].idxmax()]

print(f"Género más vendido: {genero_mas_vendido['genero']} con {genero_mas_vendido['ventas']} ventas.")
```

```python
# Estilo de Seaborn
sns.set(style="whitegrid")

# Gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='ventas', y='genero', data=ventas_por_genero_df, palette='viridis')

plt.title('Ventas Totales por Género de Videojuegos')
plt.xlabel('Ventas Totales')
plt.ylabel('Género')

plt.show()
```

Del siguiente gráfico se observa:

    - El género action domina ampliamente el mercado, con ventas muy superiores al resto.
    - Le siguen adventure y strategy, aunque a una distancia considerable.
    - Los géneros como racing presentan ventas bastante bajas en comparación.

El género más exitoso en ventas, es action con un total de 236.1905.000 ventas, destacándose ampliamente frente a otros géneros como adventure y rpg, reflejando una preferencia de los usuarios por este género.

## ¿Qué géneros tienen mejores reviews de media?

```python
#Lista para almacenar las filas
media_reviews_por_genero = []

#Se calcula la media de reviews para cada género
for genero in generos:
    # Se filtran juegos del género y se calcula la media de las calificaciones positivas
    media_reviews = df[df[genero] == True]['positive_ratings'].mean()
    # Se añade a la lista
    media_reviews_por_genero.append({'genero': genero, 'media_reviews': media_reviews})

# Se convierte la lista a un DataFrame
media_reviews_por_genero_df = pd.DataFrame(media_reviews_por_genero)

# Género con mejor review promedio
mejor_genero = media_reviews_por_genero_df.loc[media_reviews_por_genero_df['media_reviews'].idxmax()]

print(f"Género con mejor review promedio: {mejor_genero['genero']} con una promedio de {mejor_genero['media_reviews']:.2f} ratings positivos.")
```

```python
# Estilo de Seaborn
sns.set(style="whitegrid")

# Gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='media_reviews', y='genero', data=media_reviews_por_genero_df, palette='viridis')

plt.title('Promedio de los Reviews Positivos por Género')
plt.xlabel('Promedio de Reviews Positivos')
plt.ylabel('Género')

plt.show()
```

Se puede observar que los videojuegos del género action tienen en promedio la mayor cantidad de reseñas positivas con un 1580.68 de reseñas positivas.


## ¿Cuánto ha ganado el desarrollador que más juegos ha vendido?

```python
# Se agrupa por desarrollador y se suman las ventas
ganancias_desarrollador = df.groupby('developer')['sells'].sum().sort_values(ascending=False)
desarrollador_mas_vendido = ganancias_desarrollador.idxmax()
ganancia_maxima = ganancias_desarrollador.max()
print(f"\nDesarrollador que más ha vendido es: {desarrollador_mas_vendido}, Ganancias: {ganancia_maxima}")
```

```python
# Top 5 de desarrolladores
top5_desarrolladores = ganancias_desarrollador.head(5)
print("Top 5 desarrolladores por ventas totales:")
print(top5_desarrolladores)
```

```python
# Gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=top5_desarrolladores.values, y=top5_desarrolladores.index, palette="viridis")
plt.xlabel('Ventas Totales')
plt.ylabel('Desarrollador')
plt.title('Top 5 Desarrolladores por Ventas Totales')
plt.tight_layout()
plt.show()
```

La gráfica muestra que Valve es el desarrollador con mayores ventas totales en Steam. Su volumen de ventas es varias veces superior a Valve; Hidden Path Entertainment, que ocupa el segundo lugar. Además, los otros desarrolladores como PUBG Corporation, Bethesda Game Studios y Digital Extremes tienen ventas notables.

## ¿Cuál es el desarrollador (o desarrolladores) que más tipos de juego y cantidad ha desarrollado?

```python
# Desarrollador y la cantidad total de juegos
desarrolladores_info_df = df.groupby('developer').size().reset_index(name='total_games')

# Desarrollador con más juegos
max_developer = desarrolladores_info_df.loc[desarrolladores_info_df['total_games'].idxmax()]

print(f"Desarrollador con más juegos: {max_developer['developer']} "
      f"con {max_developer['total_games']} juegos.")
```

```python
# Cantidad de juegos por desarrollador
top5_desarrolladores = df['developer'].value_counts().head(5).reset_index()
top5_desarrolladores.columns = ['Desarrollador', 'Cantidad de Juegos']

# Gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top5_desarrolladores,
    y='Desarrollador',
    x='Cantidad de Juegos',
    palette='crest'
)
plt.title('Top 5 Desarrolladores por Cantidad de Juegos')
plt.xlabel('Cantidad de Juegos')
plt.ylabel('Desarrollador')
plt.tight_layout()
plt.show()
```

Se puede observar lo siguiente:

## ¿En qué mes, se han publicado más juegos?

```python
# Asegurarse de que la columna esté en formato datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Extraer el nombre del mes
df['month'] = df['release_date'].dt.month_name()

# Contar los juegos por mes
conteo_meses = df['month'].value_counts()

# Obtener el mes con más lanzamientos y su cantidad
mes_mas_juegos = conteo_meses.idxmax()
cantidad = conteo_meses.max()

print(f"Mes con más juegos publicados: {mes_mas_juegos} ({cantidad} juegos)")
```

## ¿Cuántas horas se juega de media a cada tipo de juego mensualmente?

```python
generos = ['action', 'adventure', 'rpg', 'simulation', 'strategy', 'racing']

# Convertir el dataset de formato de múltiples columnas de género a filas con un solo género
df_melted = df.melt(id_vars=['name', 'release_date', 'developer', 'achievements', 
                             'positive_ratings', 'negative_ratings', 'avg_hours_per_user', 
                             'price', 'sells', 'free_to_play', 'early_access', 'month'], 
                    value_vars=generos, var_name='genre', value_name='is_genre')

# Filtrar solo las filas donde `is_genre` es True
df_melted = df_melted[df_melted['is_genre'] == True].drop(columns=['is_genre'])
```

```python
# Agrupar por mes y género para calcular la media de horas jugadas
horas_mensuales_tipo_juego = df_melted.groupby(['month', 'genre'])['avg_hours_per_user'].mean()
print(f"\nPromedio de Horas Mensuales Jugadas por Tipo de Juego:") 

# Iterar sobre los meses
for mes in horas_mensuales_tipo_juego.index.levels[0]:  # Obtiene los nombres de los meses
    print(f"\nMes: {mes}")
    print(horas_mensuales_tipo_juego.loc[mes])
```

```python
# Configurar el gráfico
plt.figure(figsize=(12, 6))
sns.lineplot(data=horas_mensuales_tipo_juego, x="month", y="avg_hours_per_user", hue="genre", marker="o")

# Mejorar la legibilidad
plt.xticks(rotation=45)
plt.xlabel("Mes")
plt.ylabel("Horas promedio jugadas")
plt.title("Promedio de horas jugadas por género y mes")
plt.legend(title="Género")
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

Se puede concluir lo siguiente:

    - RPG y Strategy tienen picos notables en Febrero y Mayo.
    - Simulation alcanza su punto máximo en Agosto y Noviembre, lo que sugiere un mayor interés por este tipo de juegos.
    - Racing muestra una gran variabilidad, con alta actividad en Julio y Junio, pero caídas en Febrero y Septiembre.
    - Action y Adventure mantienen un comportamiento más estable a lo largo del año.

## ¿Los juegos Free to Play tienen mejores o peores reviews?

```python
# Calcular los promedios de calificaciones
reviews = df.groupby('free_to_play')[['positive_ratings', 'negative_ratings']].mean().reset_index()

# Renombrar categorías para mayor claridad
reviews['free_to_play'] = reviews['free_to_play'].map({True: 'Free to Play', False: 'Paid'})
print(f"\n Reviews de los Juegos Free to Play:") 
print(reviews)
```

```python
# Configurar el gráfico
plt.figure(figsize=(10, 6))
sns.barplot(data=reviews.melt(id_vars='free_to_play'), x='free_to_play', y='value', hue='variable')

plt.xlabel("Tipo de juego")
plt.ylabel("Promedio de calificaciones")
plt.title("Comparación de Calificaciones: Free to Play vs. Paid")
plt.legend(title="Tipo de calificación", labels=["Positivas", "Negativas"])
plt.grid(axis="y")

plt.show()
```

Los juegos Free to Play reciben más reseñas positivas, mientras que los juegos Paid tienen menos evaluaciones en general.
